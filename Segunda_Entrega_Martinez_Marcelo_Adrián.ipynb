#### **ELIMINACIÓN DE CAMPOS PARA NUEVO DF**
Para un mejor análisis y entendimiento de la correlación entre las variables se procede a crear un nuevo dataset en base al anterior. En el mismo, se eliminan aquellas variables que se consideran que no tienen impacto en la duración de la estancia de los pacientes, las cuales son: idAdmision, fechaVisita, fechaAlta, idHospital y numReadmisiones.

```python
ClinicaA= Clinica.drop(['idAdmision','fechaVisita', 'fechaAlta', 'idHospital' , 'numReadmisiones'], axis=1)
ClinicaA.head()
```

```python
ClinicaA.info()
```

```python
ClinicaA= ClinicaA.corr()
```

```python
sns.heatmap(ClinicaA, cmap = 'crest', linewidths = 0.5, linecolor = "white")
```

A partir del Heatmap anterior, no es posible identificar a primera vista cuáles son las variables que presentan una alta correlacion. Por ello, a continuación se decide representar en el Heatmap solamente los pares de varibles que presentan un grado de correlación superior al 0.30.

```python
ClinicaCor=ClinicaA.corr()
cortarget=abs(ClinicaCor["duracionEstancia"])
relevantfeatures= cortarget[cortarget>=0.30]
plt.figure(figsize=(20,10))
heatmap=sns.heatmap(ClinicaCor[abs(ClinicaCor["duracionEstancia"])>=0.30][[x for x in 
                                                                           relevantfeatures.index]], 
                    annot=True, cmap='crest', linewidths = 0.5, linecolor = "white"
                   )
plt.xticks(rotation=45) 
plt.show()
```

Mediante este Heatmap se complementa lo mencionado anteriormente, en este caso se puede apreciar cuáles son
las variables que tienen una relación más fuerte con respecto a la duración de la estancia de los pacientes en
clínica. En orden descendente se encuentra: Desorden Psicológico con una correlación de 0,51 la cual termina siendo la que presenta un grado de correlación más significativa; luego le sigue Desorden de Sangre y falta de hierro con 0,36 cada una, el atributo otros desórdenes Psíquicos con 0,35, otra variable es malnutrición con 0,32 y, por último, problemas renales con 0,3.
